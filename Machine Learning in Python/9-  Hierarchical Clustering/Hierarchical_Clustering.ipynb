# **Importing the Graphical Libraries**

```
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Importing the Dataset**

```
dataset = pd.read_csv("Mall_Customers.csv")
```

# **Creating the data vector to be used on the model**

```
X = dataset[['Annual Income (k$)', 'Spending Score (1-100)']].values
```

# **Using the dendogram in order to find the optimal number of clusters**

```
import scipy.cluster.hierarchy as sch
```

```
plt.figure(figsize = (36,18))
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
```

# **Training the Hierarchical Clustering model on the Dataset**

```
from sklearn.cluster import AgglomerativeClustering
```

```
hc = AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
```

```
y_hc = hc.fit_predict(X)
```

## ***Visualing the clusters***


```
plt.figure(figsize = (15,9))
plt.scatter(X[y_hc == 0,0],X[y_hc == 0,1], s =100, c='red',label = 'Cluster1')
plt.scatter(X[y_hc == 1,0],X[y_hc == 1,1],s =100, c='blue',label = 'Cluster2')
plt.scatter(X[y_hc == 2,0],X[y_hc == 2,1],s =100, c='green',label = 'Cluster3')
plt.scatter(X[y_hc == 3,0],X[y_hc == 3,1],s =100, c='black',label = 'Cluster4')
plt.scatter(X[y_hc == 4,0],X[y_hc == 4,1],s =100, c='brown',label = 'Cluster5')
plt.xlabel('Annual Income (K$)')
plt.ylabel('Spending Score(1-100)')
plt.title('Clusters')
plt.legend()
```
