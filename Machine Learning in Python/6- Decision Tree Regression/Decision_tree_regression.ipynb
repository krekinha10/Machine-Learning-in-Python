## **Decision Tree Regression**

```
import pandas as pd 
import numpy as np
```

```
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Importing the Dataset**

```
data = pd.read_csv('Position_Salaries.csv')
data
```

```
X = data.iloc[:,1:-1].values
X
```

```
y = data.iloc[:,-1].values
y
```

# Training the Decision Tree Regression model on the whole dataset

```
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state= 0)
```

```
regressor.fit(X,y)
```

## **Predict a new result**

```
regressor.predict([[6.5]])
```

## **Visualising the Decision tree Regression results (higher resolution)**

```
Xsmo = np.arange(data.iloc[:,1:-1].values.min(),data.iloc[:,1:-1].values.max()+0.1, step = 0.1)
Xsmo = Xsmo.reshape(Xsmo.shape[0],1)
plt.figure(figsize = (15,9))
plt.scatter(X, y, color = 'red')
plt.plot(Xsmo, regressor.predict(Xsmo), color = 'k')
plt.title('Truth or Blufff (Support Vector Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
